## Exercici 0

Fes un programa que generi una llista de números aleatòris entre 0 i 100 de longitud 25.

Després ha de generar un diccionari tipus:
```pyton
{
    'multiples2': [una llista amb els multiples de 2 de la llista aleatoria],
    'multiples3': [una llista amb els multiples de 3 de la llista aleatoria],
    'multiples5': [una llista amb els multiples de 5 de la llista aleatoria],
}
```

Dibuixa la llista i el diccionari.

````python
import random

# Genera una llista aleatòria de longitud 25 amb números entre 0 i 100
llista_aleatoria = [random.randint(0, 100) for _ in range(25)]

# Inicialitza un diccionari per emmagatzemar les llistes de múltiples
diccionari_multiples = {
    'multiples2': [],
    'multiples3': [],
    'multiples5': []
}

# Itera a través de la llista aleatòria i afegeix els números als seus respectius múltiples
for num in llista_aleatoria:
    if num % 2 == 0:
        diccionari_multiples['multiples2'].append(num)
    if num % 3 == 0:
        diccionari_multiples['multiples3'].append(num)
    if num % 5 == 0:
        diccionari_multiples['multiples5'].append(num)

print(f"Llista aleatòria: {llista_aleatoria}")

# Mostra el diccionari resultant
for key, value in diccionari_multiples.items():
    print(f"La clau '{key}' té el valor {value}")
````

## Exercici 1

Fes un programa que recorri un diccionari i intercanvii les claus pels valors.

Per exemple:
```python

    'c': 'cde',
}
{
```

````python
# Diccionari inicial
diccionari = {
    'a': 'abc',
    'b': 'bcd',
    'c': 'cde',
}

# Recorre el diccionari original i intercanvia les claus i valors
for clau, valor in list(diccionari.items()):
    diccionari[valor] = clau
    del diccionari[clau]

# Mostra el diccionari intercanviat
print(diccionari)
````

## Exercici 6

Fes un programa que a partir d'una cadena de text, retorni una llista amb els caràcters ASCII de cada lletra.

**Nota**: Per obtenir el valor ascii d'un caràcter pots fer servir la funció `ord()`.
```python
cadena = "Plou"
[80, 108, 111, 117]
```

````python
# Solució amb un 'for' tradicional
def obtenir_valors_ascii(cadena):
    valors_ascii = []
    for caracter in cadena:
        valor_ascii = ord(caracter)
        valors_ascii.append(valor_ascii)
    return valors_ascii

cadena = "Plou i fa sol."
valors_ascii = obtenir_valors_ascii(cadena)

print("Valors ASCII de cada lletra a la cadena:", valors_ascii)
````

````python
# Solució amb la funció 'map'
def obtenir_valors_ascii(cadena):
    lletres = list(cadena)
    valors_ascii = list(map(lambda x: ord(x), lletres))
    return valors_ascii

cadena = "La lluna, la bruna."
valors_ascii = obtenir_valors_ascii(cadena)

print("Valors ASCII de cada lletra a la cadena:", valors_ascii)
````

## Exercici 9

Fes un programa que retorni la matriu capgirada d'una matriu:
```python
matriu = [
    ['A', 'B', 'C'],
    ['D', 'E', 'F'],
    ['G', 'H', 'I']
]
capgirada:
    ['I', 'H', 'G']
    ['F', 'E', 'D']
    ['C', 'B', 'A']
```

````python
matriu = [
    ['A', 'B', 'C'],
    ['D', 'E', 'F'],
    ['G', 'H', 'I']
]

def girar_fila(fila):
    fila_girada = []
    for i in range(len(fila) - 1, -1, -1):
        fila_girada.append(fila[i])
    return fila_girada

def girar_matriu(matriu):
    matriu_girada = []
    for fila in range(len(matriu) - 1, -1, -1):
        fila_girada = girar_fila(matriu[fila])
        matriu_girada.append(fila_girada)
    return matriu_girada

matriu_girada = girar_matriu(matriu)
for fila in matriu_girada:
    print(fila)
````

Per exemple:
```python
matriu = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
diagonal0 = 1 * 5 * 9 = 45
diagonal1 = 3 * 5 * 7 = 105
```

````python
def multiplicar_diagonals(matriu):
    diagonal0 = 1
    diagonal1 = 1
    n = len(matriu)

    for i in range(n):
        diagonal0 = diagonal0* matriu[i][i]
        diagonal1 = diagonal1 * matriu[i][n - 1 - i]

    return diagonal0, diagonal1

matriu = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

diagonal0, diagonal1 = multiplicar_diagonals(matriu)

print("Multiplicació de la diagonal principal:", diagonal0)
print("Multiplicació de la diagonal secundària:", diagonal1)
````

## Exercici 13

Fes un programa que sumi tots els numeros d'un array ignorant els element que no siguin numèrics (int o float).
```python
array = [1, 2, 'tres', 4.5, 'cinc', 6, 'deu', 2.3]
La suma dels elements numèrics és: 15.8
```

````python
def suma_numerics(array):
    suma = 0
    for element in array:
        if isinstance(element, int) or isinstance(element, float):
            suma += element
    return suma

array = [1, 2, 'tres', 4.5, 'cinc', 6, 'deu', 2.3]
suma = suma_numerics(array)

print("La suma dels elements numèrics és:", suma)
````

````python
import json
diccionari = [
    {'nom': 'Barcelona', 'poblacio': 1620343, 'superficie': 12, 'es_capital': True}, 
    {'nom': 'Madrid', 'poblacio': 3207247, 'superficie': 667, 'es_capital': False}, 
    {'nom': 'València', 'poblacio': 791413, 'superficie': 15, 'es_capital': True}, 
    {'nom': 'Màlaga', 'poblacio': 569130, 'superficie': 11, 'es_capital': True}
]
alineacions = ["left", "center", "center", "right"]

# Retorna la longitud màxima d'una columna d'una taula
# (mira totes les files i es queda amb el valor de la que té més caràcters)
def ample_columna(taula, columna):
    longitudMaxima = 0
    for fila in taula:
        longitudFila = len(fila[columna])
        if (longitudFila > longitudMaxima):
            longitudMaxima = longitudFila
    return longitudMaxima

# Retorna un array amb l'amplada màxima de cada columna d'una taula
# (mira l'ample màxim de cada columna i fa un array amb aquestes amplades)
def amples_columnes(taula):
    amplades_columnes = []
    for columna in range(len(taula[0])):
        amplades_columnes.append(ample_columna(taula, columna))
    return amplades_columnes

# Escriu una fila, tenint en compte les alineacions de cada columna
def escriu_fila(taula, numero_fila, amplades_columnes, alineacions):
    fila = taula[numero_fila]
    txt = ""
    for numero_columna in range(0, len(fila)):
        txt_columna = fila[numero_columna]
        txt_alineat = ""
        espais_columna = amplades_columnes[numero_columna]

        # Alinea segons el tipus d'alineació
        if alineacions[numero_columna] == "left":
            txt_alineat = txt_columna.ljust(espais_columna)
        elif alineacions[numero_columna] == "center":
            txt_alineat = txt_columna.center(espais_columna)
        else: 
            txt_alineat = txt_columna.rjust(espais_columna)

        # Afegeix el separador de columnes (només si no és la última columna)
        ultima_columna = len(fila) - 1
        if numero_columna < ultima_columna:
            txt = txt + txt_alineat + " | "
        else:
            txt = txt + txt_alineat
            
    # Escriu la fila formatada        
    print(txt)

# Escriu una taula, tenint en compte les alineacions de cada columna
def escriu_taula(taula, amplades_columnes, alineacions):
    for numero_fila in range(len(taula)):
        escriu_fila(taula, numero_fila, amplades_columnes, alineacions)

# Transforma el diccionari en una taula
def diccionari_a_taula(diccionari):
    taula = []
    for element_fila in diccionari:
        fila = []
        for element_valor in element_fila.items():
            # Sempre dibuixem strings 
            # (i ens fan falta per les funcions d'alineació)
            text_valor = str(element_valor[1])
            fila.append(text_valor)
        taula.append(fila)
    return taula

# Retorna un array amb els valors màxims de dues llistes
def valors_maxims(array1, array2):
    resultat = []
    for x in range(len(array1)):
        if (array1[x] > array2[x]):
            resultat.append(array1[x])
        else:
            resultat.append(array2[x])
    return resultat

# Obtenir dues taules a partir del diccionari (titols i dades)
titols = [list(diccionari[0].keys())]
taula = diccionari_a_taula(diccionari)

# Obtenir les amplades de les columnes (a partir dels màxim dels títols i les dades)
amplades_columnes_titols = amples_columnes(titols)
amplades_columnes_taula = amples_columnes(taula)
amples_columnes = valors_maxims(amplades_columnes_titols, amplades_columnes_taula)
amplada_total = (1 + sum(amples_columnes) + len(amples_columnes) * 2)

# Escriu la taula
escriu_taula(titols, amples_columnes, alineacions)
print("-" * amplada_total)
escriu_taula(taula, amples_columnes, alineacions)
````
